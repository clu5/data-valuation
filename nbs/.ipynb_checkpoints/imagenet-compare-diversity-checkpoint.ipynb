```python
import valuation
reload(valuation)

n_components = 10

buyer_data -= np.mean(buyer_data, axis=1, keepdims=True)
buyer_cov = np.cov(buyer_data, rowvar=False)
buyer_eig_val, buyer_eig_vec = np.linalg.eig(buyer_cov)
sorted_indices = np.argsort(buyer_eig_val)[::-1]
buyer_values = buyer_eig_val[sorted_indices].real
buyer_components = buyer_eig_vec[:, sorted_indices].real


buyer_pca, buyer_cov, buyer_values, buyer_components = valuation.fit_buyer(buyer_data, n_components=n_components)
buyer_rel, buyer_div, buyer_vol = valuation.get_valuation(buyer_values, buyer_components, *valuation.project_seller(buyer_data, buyer_pca))
```

# Experiment 5

Test variability of volume with different number of samples. Compare with different datasets.

```python
# split into buyer and seller sets
seller_ratio = 0.5

indices = np.random.permutation(imagenet_embedding_matrix.shape[0])
test_size = int(imagenet_embedding_matrix.shape[0] * seller_ratio)
test_indices = indices[:test_size]
train_indices = indices[test_size:]

buyer_set = imagenet_embedding_matrix[train_indices, :]
seller_set = imagenet_embedding_matrix[test_indices, :]
```

```python
buyer_size = 100

n_components = 6
n_sizes = [100, 200, 500, 1000, 2000]
n_simulations = 10

buyer_vals2 = []
seller_vals2 = []
for i in n_sizes:
    print(f"num samples = {i}")
    buyer_data = buyer_set[:i]
    buyer_vals2.append(valuation.get_value(buyer_data, buyer_data, threshold=0.7, n_components=n_components)['volume'])
    seller_sims = []
    for n in range(n_simulations):
        indices = np.random.permutation(seller_set.shape[0])
        seller_data = seller_set[indices[:i], :]
        seller_sims.append(valuation.get_value(buyer_data, seller_data, threshold=0.7, n_components=n_components)['volume'])
    seller_vals2.append(seller_sims)
    print("Simulations complete")
    
```

```python
seller_means = []
seller_stds = []
for i in range(len(n_sizes)):
    seller_means.append(np.mean(seller_vals2[i]))
    seller_stds.append(np.std(seller_vals2[i]))

print(seller_stds)

plt.plot(n_sizes, buyer_vals2, 'k-', label="Buyer Data")
plt.errorbar(n_sizes, seller_means, seller_stds, fmt='o', capsize=5,  label="Seller Data Samples")
plt.xlabel("Num samples")
plt.ylabel("Volume")
plt.yscale('log')
plt.legend()
plt.show()
```
